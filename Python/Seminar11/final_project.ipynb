#### f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
import sympy 
import numpy as np
from sympy.plotting import plot
from sympy import symbols
```

```python
# определяем корни с помощью функции solveset()
x = symbols('x')
sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
# вот такой интересный результат получился.
# и судя по графику ниже - корней у уравнения много. Соответственно функция имеет бесконечное число интервалов, 
# где она возрастает или убывает. И вершину определить можно, только на каком-то заданном интервале   
```

```python
# График функции
x = symbols('x')
plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100, 0.0001))
```

```python
x = np.arange(-10, 10)
f = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

positiv_lst = []
negativ_lst = []

for i in x:
    result = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if result > 0:
        positiv_lst.append(i)
    else:
        negativ_lst.append(i)

print(f'Промежутки, на которых f > 0: {positiv_lst[-1]} --> {positiv_lst[0]}')
print(positiv_lst)
print(f'Промежутки, на которых f < 0:{negativ_lst[0]} --> {negativ_lst[-1]}')
print(negativ_lst)
```
